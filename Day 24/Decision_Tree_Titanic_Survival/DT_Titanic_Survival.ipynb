```python
import pandas as pd
import numpy as np #storing as array
from sklearn import tree
from sklearn import preprocessing 
# used for finding missing values, text into numrical
```

```python
train = pd.read_csv(r'C:\Users\mansi\Documents\LetsUpgrade\Day 24\Decision_Tree_Titanic_Survival\train.csv')
train.head(3)
```

```python
#finding missing values
train.isna().sum() #Checking NA/Null values  #cabin having NA values.
```

```python
train["Age"].mean()
```

```python
new_age=np.where(train["Age"].isnull(),32,train["Age"])

train["Age"]=new_age

train['Sex'] = train['Sex'].map({'male':1, 'female': 0})

train.head()
```

```python
data1=pd.DataFrame([train["Sex"],train["Age"],train["Fare"]]).T
```

```python

treemodel=tree.DecisionTreeClassifier(max_depth=6)

```

```python
treemodel.fit(X=data1,y=train["Survived"])
```

```python
treemodel.score(X=data1,y=train["Survived"])
```

```python
# Prediction of Survival Using Age,Sex,Fare

test=pd.read_csv(r"C:\Users\mansi\Documents\LetsUpgrade\Day 24\Decision_Tree_Titanic_Survival\test.csv")

test["Age"].mean()

33.23741007194245

new_age2=np.where(test["Age"].isnull(),33,test["Age"])

test["Age"]=new_age2
test['Sex'] = test['Sex'].map({'male':1, 'female': 0})

test.head()
```

```python
data2=pd.DataFrame([test["Sex"], test["Age"],test["Fare"]]).T

test_preds=treemodel.predict(X=data2)

predicted_output=pd.DataFrame({"PassengerId":test["PassengerId"],"Survived":test_preds})

predicted_output.to_csv("output.csv",index=False);
```
